```python
#Choose one variable and plot that variable four different ways.
sns.set(color_codes=True)
DSWeatherData['MinTemp'] = DSWeatherData['MinTemp'].fillna(0)
DSWeatherData['MaxTemp'] = DSWeatherData['MaxTemp'].fillna(0)
plt.figure(figsize=(15,8))
sns.distplot(DSWeatherData.MinTemp, bins=100, hist=False); #rug=True, 
```

```python
#kernel density estimate
sns.kdeplot(DSWeatherData.MinTemp)
sns.kdeplot(DSWeatherData.MinTemp, bw=.5, label="bw: 0.5")
sns.kdeplot(DSWeatherData.MinTemp, bw=2, label="bw: 2")
plt.legend();
```

```python
sns.distplot(DSWeatherData.MinTemp, kde=False, fit=stats.gamma);
```

```python
sns.rugplot(DSWeatherData.MinTemp, color=".4", linewidth=1);
```

```python
#Choose two continuous variables, and plot them three different ways.
#Choosing MinTemp and MaxTemp
sns.jointplot(x=DSWeatherData.MinTemp, y=DSWeatherData.MaxTemp, color='red', kind="hex");
```

```python
sns.kdeplot(DSWeatherData.MinTemp.head(100), DSWeatherData.MaxTemp.head(100))
sns.rugplot(DSWeatherData.MinTemp.head(100), color="g")
sns.rugplot(DSWeatherData.MaxTemp.head(100), vertical=True);
```

```python
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(DSWeatherData.MinTemp.head(100), DSWeatherData.MaxTemp.head(100), cmap=cmap, n_levels=60, shade=True);
```
