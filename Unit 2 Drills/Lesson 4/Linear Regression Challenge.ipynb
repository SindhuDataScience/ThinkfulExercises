```python
df = df.fillna(0)
df.count()
```

```python
df['PopulationSquare'] = df['Population'] * df['Population']
df.head(5)
```

```python
df['MurderCat'] = np.where(df['Murder and nonnegligent manslaughter'] > 0, 1, 0)
df['RobberyCat'] = df.apply(lambda row: 1 if row.Robbery > 0 else 0, axis=1)
```

```python
#Multiple Linear Regression
#𝑃𝑟𝑜𝑝𝑒𝑟𝑡𝑦𝑐𝑟𝑖𝑚𝑒=𝛼+𝑃𝑜𝑝𝑢𝑙𝑎𝑡𝑖𝑜𝑛+𝑃𝑜𝑝𝑢𝑙𝑎𝑡𝑖𝑜𝑛2+𝑀𝑢𝑟𝑑𝑒𝑟+𝑅𝑜𝑏𝑏𝑒𝑟𝑦
from sklearn import linear_model
Xs = pd.DataFrame(data=df,columns=['Population','PopulationSquare','MurderCat','RobberyCat'])
y = df['Property crime'].values.reshape(-1,1)
reg = linear_model.LinearRegression()
reg.fit(Xs, y)
print("The linear model is: PropertyCrime = {:.5} + {:.5}*Population + {:.5}*PopulationSquare + {:.5}*Murder + {:.5}*Robbery".format(reg.intercept_[0], reg.coef_[0][0], reg.coef_[0][1], reg.coef_[0][2], reg.coef_[0][3]))
```

```python
import statsmodels.api as sm
X = np.column_stack((df['Population'], df['PopulationSquare'], df['MurderCat'], df['RobberyCat']))
y = df['Property crime']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```
