```python
df = pd.get_dummies(df, columns=['WindGustDir','WindDir9am','WindDir3pm'])
```

```python
df.head()
```

```python
from sklearn import tree
from IPython.display import Image
import pydotplus

df_cols = df.drop(columns=['RainToday','RainTomorrow'])

decision_tree = tree.DecisionTreeClassifier(
    criterion='entropy',
    max_features=5,
    max_depth=4,
    random_state = 0
)
decision_tree.fit(df_cols, df['RainToday'])
target_pred_dt = decision_tree.predict(df_cols)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(decision_tree, df_cols, df['RainToday'], cv=10)
#~3 sec to run
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(df['RainToday'], target_pred_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfclassifier = RandomForestClassifier(
    n_estimators=10, 
    criterion = 'entropy', 
    random_state=1)
rfclassifier.fit(df_cols, df['RainToday'])
target_pred_rf = rfclassifier.predict(df_cols)
```

```python
cross_val_score(rfclassifier, df_cols, df['RainToday'], cv=10)
#~10 secs to run
```

```python
confusion_matrix(df['RainToday'], target_pred_rf)
```
