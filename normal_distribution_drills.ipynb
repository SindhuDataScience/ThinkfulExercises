DRILL - Descriptive statistics and Normality

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#exponential, geometric, laplace, rayleigh, logseries, poisson
exponential = np.random.exponential(0.12, 100)
geometric = np.random.geometric(0.25, 100)
laplace = np.random.laplace(5, 0.25, 100)
rayleigh = np.random.rayleigh(0.44, 100)
logseries = np.random.logseries(0.3, 100)
poisson = np.random.poisson(3, 100)

plt.title('Exponential distribution')
plt.hist(exponential, color='lightgrey')
plt.axvline(exponential.mean(), linestyle='solid')
plt.axvline(exponential.mean() + exponential.std(), linestyle='dotted')
plt.axvline(exponential.mean() - exponential.std(), linestyle='dotted')
plt.show()

plt.title('Geometric distribution')
plt.hist(geometric, color='lightgrey')
plt.axvline(geometric.mean(), linestyle='solid')
plt.axvline(geometric.mean() + geometric.std(), linestyle='dotted')
plt.axvline(geometric.mean() - geometric.std(), linestyle='dotted')
plt.show()

plt.title('Laplace distribution')
plt.hist(laplace, color='lightgrey')
plt.axvline(laplace.mean(), linestyle='solid')
plt.axvline(laplace.mean() + laplace.std(), linestyle='dotted')
plt.axvline(laplace.mean() - laplace.std(), linestyle='dotted')
plt.show()

plt.title('Rayleigh distribution')
plt.hist(rayleigh, color='lightgrey')
plt.axvline(rayleigh.mean(), linestyle='solid')
plt.axvline(rayleigh.mean() + rayleigh.std(), linestyle='dotted')
plt.axvline(rayleigh.mean() - rayleigh.std(), linestyle='dotted')
plt.show()

plt.title('Logseries distribution')
plt.hist(logseries, color='lightgrey')
plt.axvline(logseries.mean(), linestyle='solid')
plt.axvline(logseries.mean() + logseries.std(), linestyle='dotted')
plt.axvline(logseries.mean() - logseries.std(), linestyle='dotted')
plt.show()

plt.title('Poisson distribution')
plt.hist(poisson, color='lightgrey')
plt.axvline(poisson.mean(), linestyle='solid')
plt.axvline(poisson.mean() + poisson.std(), linestyle='dotted')
plt.axvline(poisson.mean() - poisson.std(), linestyle='dotted')
plt.show()
```

Additionally:

Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.
Add them together to create a third variable.
Graph the third variable using a histogram.
Compute the mean and standard deviation and plot them as vertical lines on the histogram.
Evaluate the descriptive statistics against the data.

```python
normal1 = np.random.normal(5,0.5,1000)
normal2 = np.random.normal(10,1,1000)
plt.hist(normal1)
plt.show()

plt.hist(normal2)
plt.show()
```

```python
normal3 = normal1 + normal2
plt.hist(normal3, color='lightgrey')
plt.axvline(normal3.mean(), linestyle='solid')
plt.axvline(normal3.mean() + normal3.std(), linestyle='dotted')
plt.axvline(normal3.mean() - normal3.std(), linestyle='dotted')
plt.show()
```

```python
normaltst1 = np.random.normal(5,0.5,10000)
normaltst2 = np.random.normal(10,1,10000)
normaltst3 = np.sort(normaltst1) + np.sort(normaltst2)
normaltst4 = normaltst1 + normaltst2
#df = pd.DataFrame()
#df['n1'] = normaltst1
#df['n2'] = normaltst2
#df['n3'] = normaltst3
#print(normaltst1)
#print(normaltst2)
#print(normaltst3)
#df = pd.DataFrame(normaltst1, normaltst2, normaltst3)
#df
plt.hist(normaltst1, color='lightgrey')
plt.axvline(normaltst1.mean(), linestyle='solid')
plt.axvline(normaltst1.mean() + normaltst1.std(), linestyle='dotted')
plt.axvline(normaltst1.mean() - normaltst1.std(), linestyle='dotted')
plt.show()

plt.hist(normaltst2, color='lightgrey')
plt.axvline(normaltst2.mean(), linestyle='solid')
plt.axvline(normaltst2.mean() + normaltst2.std(), linestyle='dotted')
plt.axvline(normaltst2.mean() - normaltst2.std(), linestyle='dotted')
plt.show()

plt.hist(normaltst3, color='lightgrey')
plt.axvline(normaltst3.mean(), linestyle='solid')
plt.axvline(normaltst3.mean() + normaltst3.std(), linestyle='dotted')
plt.axvline(normaltst3.mean() - normaltst3.std(), linestyle='dotted')
plt.show()

plt.hist(normaltst4, color='lightgrey')
plt.axvline(normaltst4.mean(), linestyle='solid')
plt.axvline(normaltst4.mean() + normaltst4.std(), linestyle='dotted')
plt.axvline(normaltst4.mean() - normaltst4.std(), linestyle='dotted')
plt.show()

print('Normaltst1 statistics:',normaltst1.mean(), normaltst1.std())
print('Normaltst2 statistics:',normaltst2.mean(), normaltst2.std())
print('Normaltst3 statistics:',normaltst3.mean(), normaltst3.std())
print('Normaltst4 statistics:',normaltst4.mean(), normaltst4.std())
```

Looking at the above graphs, adding two lists of normally distributed data points will yield another set of data that is normall distributed as well (using both sorted lists and unsorted lists). Mean and SD of the new list are the summed means and SDs of the parent lists.
